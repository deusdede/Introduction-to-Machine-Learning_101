# K Nearest Neighbors using Iris

```python
from __future__ import print_function
import os
import numpy as np
```

```python
import pandas as pd

# Import the data using the file path
# filepath = os.sep.join(data_path + ['Orange_Telecom_Churn_Data.csv'])
data = pd.read_csv('Iris_Data.csv')
```

```python
data.sample(10)
```

```python
data.columns
```

```python
data['species_pred'] = data['species']
```

```python
data.sample(10)
```

```python
data['species_pred'] = np.where(data.species_pred=="Iris-virginica", 0, data.species_pred)
data['species_pred'] = np.where(data.species_pred=="Iris-setosa", 1, data.species_pred)
data['species_pred'] = np.where(data.species_pred=="Iris-versicolor", 2, data.species_pred)
```

```python
data.sample(10)
```

```python
# Get a list of all the columns that don't contain the label
x_cols = [x for x in data.columns if (x != 'species' and x != 'species_pred')]

# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['species_pred']
```

```python
print(y_data.tolist())
```

```python
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor(n_neighbors=1,weights='distance')

knn = neigh.fit(X_data, y_data)
y_pred = knn.predict(X_data)
y_pred = y_pred.astype(int)
```
